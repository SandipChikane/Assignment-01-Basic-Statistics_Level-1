```python
data=pd.read_csv('Q9_b.csv')
```

```python
data.head()
```

```python
#Inference for data
data.describe()
```

```python
#Skewness of data
print("Skew of SP:-",data.SP.skew())
print("Skew of WT:-",data.WT.skew())
```

```python
#Kurtosis of data
from scipy.stats import kurtosis
print("Kurtosis of SP:-",kurtosis(data.SP))
print("Kurtosis of WT:-",kurtosis(data.WT))
```

```python
#Range for SP & WT
print("Range of SP:-",data.SP.max()-data.SP.min())
print("Range of WT:-",data.WT.max()-data.WT.min())
```

```python
#Variance for SP & WT
print("Variance for SP :-",data.SP.var())
print("Variance for WT :-",data.WT.var())
```

```python
#Mode of SP & WT
import statistics
print("Mode of SP :-",statistics.mode(data.SP))
print("Mode of WT :-",statistics.mode(data.WT))
```

```python
#Median of SP & WT
print("Median of SP :-",statistics.median(data.SP))
print("Median of WT :-",statistics.median(data.WT))
```
